```python
import numpy as np
fourmis_pos=np.load('result/boite/location_seul.npy', allow_pickle=True)
fourmis_pos=np.array(fourmis_pos,dtype=object)
max([len(fourmis_pos[i]) for i in range(len(fourmis_pos))])
```

```python
def track_one(pos,fourmis_matrix,x,y,i=0):
    if i>len(fourmis_matrix)-20:
        print('Fini')
        return
    mat=fourmis_matrix[i:i+30]
    #print('i=',i)
    for t in range(1,30):
        #print('t=',t)
        for k in range (len(mat[t])):
            point=mat[t][k]
            if np.linalg.norm(point - pos)<t*10:
                #print('pos found at frame',i+t,'pos=',point)
                x.append(point[0])
                y.append(point[1])
                track_one(pos,fourmis_matrix,x,y,i+t)
                return x,y
    print('perdu a',i)
    return x,y
```
